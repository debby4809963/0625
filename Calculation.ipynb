```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from matplotlib import cm


def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,  #透明
                linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```

```python
from sklearn import datasets, cluster, metrics
iris = datasets.load_iris()
iris_X=iris.data

result=cluster.KMeans(n_clusters=3).fit(iris_X)
result.labels_

#計算silhouette score,越接近1分群結果越好
silhouette_avg = metrics.silhouette_score(iris_X, result.labels_)
print(silhouette_avg)
```

```python
from sklearn import datasets, cluster, metrics
import matplotlib.pyplot as pt
iris = datasets.load_iris()
iris_X=iris.data

ks=range(2,11)
silhouette_avgs=[]

for n in ks :
    result=cluster.KMeans(n_clusters=n).fit(iris_X)
    silhouette_avg = metrics.silhouette_score(iris_X, result.labels_)
    silhouette_avgs.append(silhouette_avg)
    
pt.bar(ks,silhouette_avgs)
pt.show()
```

```python
from sklearn import datasets
from sklearn.cross_validation import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt 
boston=datasets.load_boston()
#print(boston.DESCR)
#print(boston.data)
#print(boston.target)
#CRIM(犯罪率) ZN(房屋大於25000ft比率) INDUS(住宅比率)
#CHAS(有無臨河) NOX(空屋比率) RM(房間數) AGE(自有住宅比率)
#DIS(離市中心距離) RAD(離高速公路距離) TAX(房屋稅率)
#PTRATIO(小學老師比率) B(黑人比率) LSTAT(低收入比率) MEDV(受雇者收入)
lr=linear_model.LinearRegression()
predicted = cross_val_predict(lr,boston.data,boston.target,cv=10)

y=boston.target
plt.figure
plt.scatter(y,predicted)
plt.plot([y.min(),y.max()],[y.min(),y.max()],'k--',lw=4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from sklearn.linear_model import LinearRegression
from sklearn.isotonic import IsotonicRegression
from sklearn.utils import check_random_state
n=100
x=np.arange(n)
rs=check_random_state(0)
y=rs.randint(-50,50,size=(n,))+50.*np.log(1+np.arange(n))
ir=IsotonicRegression()
y_=ir.fit_transform(x,y)
lr=LinearRegression()
lr.fit(x[:,np.newaxis],y)
segments=[[[i,y[i]],[i,y_[i]]] for i in range(n)]
lc=LineCollection(segments,zorder=0)
lc.set_array(np.ones(len(y)))
lc.set_linewidths(0.5*np.ones(n))
fig=plt.figure()
plt.plot(x,y,'r.',markersize=12)
plt.plot(x,y_,'g.-',markersize=12)
plt.plot(x,lr.predict(x[:,np.newaxis]),'b-')
plt.gca().add_collection(lc)
plt.legend(('Data','IsotonicFit','Linear Fit'),loc='lower right')
plt.title('Isotonic regression')
plt.show()
```

```python
from sklearn import datasets, linear_model,svm
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
diabetes = datasets.load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(diabetes.data,diabetes.target,test_size=0.2)
svc=svm.SVR(kernel='linear',C=100,gamma=0.001)
svc.fit(X_train,y_train)
predicted=svc.predict(X_test)
plt.scatter(y_test,predicted)
y=diabetes.target
plt.plot([y.min(),y.max()],[y.min(),y.max()],'k--',lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.show()
```
